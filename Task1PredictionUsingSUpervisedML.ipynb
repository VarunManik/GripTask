Import Required Python Libaries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Reading Data

```python
data=pd.read_csv("http://bit.ly/w-data")
print("Data IMPORTED Sucessfully")

data.head(15)
```

Plot the given Data Hours VS Percentage

```python
data.plot(x="Hours",y="Scores",style="o")
plt.title("Hours VS Percentage")
plt.xlabel("Hours Studied")
plt.ylabel("Percentage Score")
plt.show()
```

```python
data.describe()
```

```python
x=data.iloc[:,[0]].values
y=data.iloc[:,[1]].values
x.shape,y.shape
```

Spliting the Training and Testing data from given data

```python
from sklearn.model_selection import train_test_split
x_train,x_valid,y_train,y_valid=train_test_split(x,y,random_state=2,test_size=0.2) 
x_train.shape, x_valid.shape, y_train.shape, y_valid.shape
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
print("TrainingCompleted")
```

Displaying the dataframe the actual and predicted values

```python
predic=model.predict(x_valid)
diff=y_valid-predic
pd.DataFrame(np.c_[y_valid,predic,diff] ,columns=['Actual','Predicted','Difference'])
```

Plot of training data set
blue line: predicted model
red dots: actual values

```python
plt.scatter(x_train, y_train, color="red")
plt.plot(x_train,model.predict(x_train),color="blue")
plt.title("Hours VS Scores")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.show()
```

Plot of model data set
blue line: predicted model
red dots: actual values

```python
plt.scatter(x_valid, y_valid, color="red")
plt.plot(x_valid,model.predict(x_valid),color="blue")
plt.title("Hours VS Scores")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.show()
```

Checking Error

```python
from sklearn import metrics
print("Mean Abslout Error : ",metrics.mean_absolute_error(y_valid,predic))
```

Percentage Prediction 

```python
hours=9.25
predicition = model.predict([[hours]])[0]
print("No of Hours = {}".format(hours),"Predicted Score = {}%".format(predicition))
```
